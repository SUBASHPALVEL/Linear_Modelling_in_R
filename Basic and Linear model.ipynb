# Test and train split

```r
set.seed(0)
split=sample.split(df,SplitRatio=0.8)
```

```r
training_set=subset(df,split==TRUE)
test_set=subset(df,split==FALSE)
```

# Linear Model

```r
lm_a=lm(price~.,data=training_set)
train_a=predict(lm_a,training_set)
test_a=predict(lm_a,test_set)
```

# Ridge regression in R

```r
x=model.matrix(price~.,data=df)[,-1]
y=df$price
```

```r
grid=10^seq(10,-2,length=100)
grid
```

```r
library("glmnet")
```

```r
lm_ridge=glmnet(x,y,alpha=0,lambda=grid)
summary(lm_ridge)
```

```r
cv_fit=cv.glmnet(x,y,alpha=0,lambda=grid)
plot(cv_fit)
```

```r
opt_lambda=cv_fit$lambda.min
```

```r
tss=sum((y-mean(y))^2)
```

```r
y_a =predict(lm_ridge,s=opt_lambda,newx=x)
rss=sum((y_a-y)^2)
```

```r
reg=1-rss/tss
```

# Lesso regression in R

```r
lm_lesso=glmnet(x,y,alpha=1,lambda=grid)
summary(lm_lesso)
```

```r
cv_fit=cv.glmnet(x,y,alpha=1,lambda=grid)
plot(cv_fit)
```

```r
opt_lambda=cv_fit$lambda.min
tss=sum((y-mean(y))^2)
y_a =predict(lm_lesso,s=opt_lambda,newx=x)
rss=sum((y_a-y)^2)
```

```r
reg=1-rss/tss
```
